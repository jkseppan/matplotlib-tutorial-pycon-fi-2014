## Goal of this talk

<p style="margin: 30px 3em 30px 3em; font-size: 110%">**[Matplotlib](http://matplotlib.org/)** is a very powerful and popular plotting library for Python. It is easy to get started with plotting, but those easy plots tend to all look alike, and the default settings do not fit every plot's exact needs. In this talk I will show some of my tricks to figure out what can be customized and how to get it done, and on the way will advertise some of the things that are possible with matplotlib. The talk will involve code samples and some live experimentation in order to show the workflow.
</p>

This is an **IPython notebook**, and I highly recommend IPython for developing plots. To initialize matplotlib, use the "magic command" `%matplotlib backend`, or `%pylab backend`. The difference is that `%matplotlib` merely sets up the plotting, `%pylab` imports a large number of utilities into the interactive namespace. We're going to do the latter here, but do note that this pollutes the namespace and can get *very annoying* when writing larger programs, so if you are going to reuse anything you should use namespaces properly, e.g.

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

and then use `np.linspace` instead of `linspace`, `plt.figure` instead of `figure`, etc.

The "backend" mentioned above refers to the variety of ways to output a plot. The "inline" backend used here is a good counterpart to the IPython notebook, and in recent versions of matplotlib there is an "nbagg" backend that adds interactive features. There are backends for saving in several different file formats, and interactive backends for most GUI toolkits supported by Python.

## The most important functions in pyplot

* Format the figure: `subplots`, `subplots_adjust`, `tight_layout`
* Plot some data: `plot`, `imshow`, `hist`, `boxplot`
  * Support from numpy: `linspace`, `meshgrid`
* Explain the plot: `legend`, `text`, `xlabel`, `ylabel`, `title`
* Fine tune the formatting
  * `xlim`, `ylim`
  * `gca`, `gcf`, `getp`, `setp`
* Get the plot out there
  * `show`, `close` (depends on backend and interactivity)
  * `savefig`
  
---

## The most important properties of Artists

These vary from one Artist to another, and there are many more.

* For the legend: `label`
* Hide unwanted artists: `visible`
* Color: `facecolor`, `edgecolor`, `color`, `alpha`
* Text: `fontproperties`, `size`
* Lines: `linestyle`, `linewidth`
* Markers: `marker`, `markeredgecolor`, `markerfacecolor`, `markeredgewidth`, `markersize`
* Location: `transform` (see `ax.transData` and friends)
* What is on top of what: `zorder`

## Plotting project:  the normal distribution

Imagine you're a math teacher and want to explain the [normal distribution](http://en.wikipedia.org/wiki/Normal_distribution) with plots. The `scipy.stats.norm` object includes all the numerics needed. We should plot the density function and the cumulative distribution function.

````python
import scipy.stats
norm = scipy.stats.norm
````

````python
plot(norm.pdf(linspace(-4, 4, 300)))
````

We can start from a Hello World style plot, but we'll need two subplots. The simple way to create the objects is the `plt.subplots` function.

````python
%load normal1.py
````

````python
fig, (ax1, ax2) = subplots(nrows=2)
x = linspace(-4, 4, 300)
ax1.plot(x, norm.pdf(x), lw=2)
ax2.plot(x, norm.cdf(x), lw=2)
````

````python
fig, (ax1, ax2) = subplots(nrows=2, gridspec_kw=dict(hspace=0.4))

x = linspace(-4, 4, 300)
pdf = norm.pdf(x)
cdf = norm.cdf(x)

def setup_axis(ax):
    ax.set_xlim(-4.5, 4.5)
    ax.xaxis.set_tick_params(labelsize=14, direction='out')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_tick_params(labelsize=14, direction='out')
    ax.yaxis.set_ticks_position('left')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_position(('outward', 10))
    ax.spines['left'].set_position(('outward', 10))

setup_axis(ax1); setup_axis(ax2)
ax1.plot(x, norm.pdf(x), lw=2)
ax1.set_ylim(-0.05, 0.45)
ax2.plot(x, norm.cdf(x), 'g', lw=2)
ax2.set_ylim(-0.05, 1.05)
````

````python
%load normal3.py
# This has the fill, and some horizontal and vertical lines
# (hlines, vlines).
````

````python
fig, (ax1, ax2) = subplots(nrows=2, gridspec_kw=dict(hspace=0.4))

x = linspace(-4, 4, 300)
pdf = norm.pdf(x)
cdf = norm.cdf(x)
idx = 170

def setup_axis(ax):
    ax.set_xlim(-4.5, 4.5)
    ax.set_xticks(range(-4, 5))
    ax.xaxis.set_tick_params(labelsize=14, direction='out')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_tick_params(labelsize=14, direction='out')
    ax.yaxis.set_ticks_position('left')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_position(('outward', 10))
    ax.spines['left'].set_position(('outward', 10))

setup_axis(ax1); setup_axis(ax2)
ax1.plot(x, norm.pdf(x), lw=2)
ax1.fill_between(x[:idx], pdf[:idx], color='g')
ax1.set_ylim(-0.05, 0.45)
ax1.set_ylabel('density', size=20)
ax1.spines['left'].set_bounds(0, max(pdf))
ax1.spines['bottom'].set_bounds(-4, 4)

ax2.plot(x, norm.cdf(x), 'g', lw=2)
ax2.grid(linestyle=':')
ax2.set_ylim(-0.05, 1.05)
ax2.set_ylabel('cumulative', size=20)
ax2.spines['left'].set_bounds(0, 1)
ax2.spines['bottom'].set_bounds(-4, 4)
vline=ax2.vlines(x[idx-1], 0, 2.65, lw=2, linestyle='--')
vline.set_clip_on(False)
ax2.hlines(cdf[idx-1], -5, x[idx-1], lw=2, linestyle='--')
````

````python
fig, (ax1, ax2) = subplots(nrows=2, gridspec_kw=dict(hspace=0.4))

x = linspace(-4, 4, 300)
pdf = norm.pdf(x)
cdf = norm.cdf(x)
idx = 170

def setup_axis(ax):
    ax.set_xlim(-4.5, 4.5)
    ax.set_xticks(range(-4, 5))
    ax.xaxis.set_tick_params(labelsize=14, direction='out')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_tick_params(labelsize=14, direction='out')
    ax.yaxis.set_ticks_position('left')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_position(('outward', 10))
    ax.spines['left'].set_position(('outward', 10))

setup_axis(ax1); setup_axis(ax2)
ax1.plot(x, norm.pdf(x), lw=2)
ax1.fill_between(x[:idx], pdf[:idx], color='g')
ax1.set_ylim(-0.05, 0.45)
ax1.set_ylabel('density', size=20)
ax1.spines['left'].set_bounds(0, max(pdf))
ax1.spines['bottom'].set_bounds(-4, 4)

ax2.plot(x, norm.cdf(x), 'g', lw=2)
ax2.grid(linestyle=':')
ax2.set_ylim(-0.05, 1.05)
ax2.set_ylabel('cumulative', size=20)
ax2.set_yticks(norm.cdf([-2,-1,0,1,2]))
ax2.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
ax2.spines['left'].set_bounds(0, 1)
ax2.spines['bottom'].set_bounds(-4, 4)
vline=ax2.vlines(x[idx-1], 0, 2.65, lw=2, linestyle='--')
vline.set_clip_on(False)
ax2.hlines(cdf[idx-1], -5, x[idx-1], lw=2, linestyle='--')

ax1.text(0.1, 0.7, r'$\frac{1}{\sqrt{2\pi}}e^{\frac{-x^2}{2}}$',
         fontsize=30, transform=ax1.transAxes)
````
